# Geographical Locations of the Data

```python
import pandas as pd
import geopandas as gpd
import folium
from folium.plugins import MarkerCluster

# Load your dataset (make sure to adjust the file path as necessary)
df = pd.read_csv("C:/Users/alhas/Desktop/Jupiter Lab Project/SHS-Data.csv")

# Define the correct column names for longitude and latitude
lon_col = "gps-Longitude"
lat_col = "gps-Latitude"  

# Print column names to verify
print("Columns in the DataFrame:")
print(df.columns)

# Verify that the specified columns exist in the DataFrame
if lon_col not in df.columns or lat_col not in df.columns:
    raise ValueError(f"One or both of the specified columns ({lon_col}, {lat_col}) are not in the DataFrame.")

# Remove rows with NaN values in coordinate columns
df_clean = df.dropna(subset=[lon_col, lat_col])

# Print the number of rows removed
print(f"Rows in original DataFrame: {len(df)}")
print(f"Rows in cleaned DataFrame: {len(df_clean)}")
print(f"Number of rows removed due to NaN coordinates: {len(df) - len(df_clean)}")

# Convert the cleaned DataFrame to a GeoDataFrame
gdf = gpd.GeoDataFrame(df_clean, geometry=gpd.points_from_xy(df_clean[lon_col], df_clean[lat_col]))

# Create a map centered on the mean coordinates of your data
center_lat = gdf[lat_col].mean()
center_lon = gdf[lon_col].mean()
m = folium.Map(location=[center_lat, center_lon], zoom_start=10)

# Add a marker cluster layer
marker_cluster = MarkerCluster().add_to(m)

# Add markers for each point
for idx, row in gdf.iterrows():
    folium.Marker(
        location=[row[lat_col], row[lon_col]],
        popup=f"ID: {idx}",  # You can customize this popup with other columns
    ).add_to(marker_cluster)

# Add layer control
folium.LayerControl().add_to(m)

# Display the map
m
```

# Participants Age Distribution across the Study Location

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (update the path accordingly)
df = pd.read_csv(r"C:\Users\alhas\Desktop\Jupiter Lab Project\SHS-Data.csv")

# Define the age bins and their labels
age_bins = [0, 18, 35, 50, 65, 100]  # These can be adjusted as needed
age_labels = ['0-18', '19-35', '36-50', '51-65', '66+']

# Categorize the 'respondent_age' column into these age bins
df['age_group'] = pd.cut(df['respondent_age'], bins=age_bins, labels=age_labels, right=False)

# Drop any rows where 'respondent_age' or 'age_group' is NaN
df = df.dropna(subset=['respondent_age'])

# Count the number of respondents in each age group
age_group_counts = df['age_group'].value_counts(sort=False)

# Set the style and color palette for better aesthetics
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Create a bar plot for age group counts
bars = sns.barplot(x=age_group_counts.index, y=age_group_counts.values, palette='bright')

# Set titles and labels
plt.title('Respondent Age Group Distribution', fontsize=18)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Add data labels (counts) to the bars
for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.0f'),
                  (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                  ha='center', va='bottom', fontsize=12, color='black')

# Adjust layout for a cleaner look
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (update the path accordingly)
df = pd.read_csv(r"C:\Users\alhas\Desktop\Jupiter Lab Project\SHS-Data.csv")

# Fill missing values in 'visited_health_facility'
df['visited_health_facility'] = df['visited_health_facility'].fillna('No')

def plot_crosstab(crosstab, title, xlabel, ylabel):
    """Function to create a bar plot from a crosstab."""
    plt.figure(figsize=(10, 6))
    melted = crosstab.iloc[:-1, :-1].reset_index().melt(id_vars='visited_health_facility')
    bars = sns.barplot(data=melted, x=xlabel, y='value', hue='visited_health_facility', palette='pastel')
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.xticks(rotation=45)
    plt.legend(title='Visited Health Facility')

    # Add data labels
    for bar in bars.patches:
        bars.annotate(format(bar.get_height(), '.0f'), 
                      (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                      ha='center', va='bottom', fontsize=12)

    plt.show()

# Crosstab for visited_health_facility and visit_frequency
crosstab_visited_frequency = pd.crosstab(
    df['visited_health_facility'], 
    df['visit_frequency'], 
    margins=True, 
    margins_name="Total",
    dropna=False
)
print("Cross-Tabulation of Visited Health Facility by Visit Frequency:")
print(crosstab_visited_frequency)
plot_crosstab(crosstab_visited_frequency, 'Utilization of Health Facilities by Visit Frequency', 
               'visit_frequency', 'Total Visits')

# Crosstab for visited_health_facility and facility_most_visited
crosstab_most_visited = pd.crosstab(
    df['visited_health_facility'], 
    df['facility_most_visited'], 
    margins=True, 
    margins_name="Total",
    dropna=False
)
print("\nCross-Tabulation of Visited Health Facility by Most Visited Facility:")
print(crosstab_most_visited)
plot_crosstab(crosstab_most_visited, 'Most Visited Health Facilities', 
               'facility_most_visited', 'Number of Visits')

# Crosstab for visited_health_facility and other_facility
crosstab_other_facility = pd.crosstab(
    df['visited_health_facility'], 
    df['other_facility'], 
    margins=True, 
    margins_name="Total",
    dropna=False
)

# Crosstab for visited_health_facility and reason_for_visit
crosstab_reason_visit = pd.crosstab(
    df['visited_health_facility'], 
    df['reason_for_visit'], 
    margins=True, 
    margins_name="Total",
    dropna=False
)
print("\nCross-Tabulation of Visited Health Facility by Reason for Visit:")
print(crosstab_reason_visit)
plot_crosstab(crosstab_reason_visit, 'Reasons for Visiting Health Facilities', 
               'reason_for_visit', 'Number of Visits')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (adjust the file path if necessary)
df = pd.read_csv("C:/Users/alhas/Desktop/Jupiter Lab Project/SHS-Data.csv")

# Check if the 'barriers_to_access' column exists
if 'barriers_to_access' not in df.columns:
    print("Error: 'barriers_to_access' column not found.")
else:
    # Group data by district and barriers, and calculate the count
    district_barriers = df.groupby(['district', 'barriers_to_access']).size().reset_index(name='count')

    # List of districts
    districts = district_barriers['district'].unique()

    # Loop through each district and create a separate plot
    for district in districts:
        plt.figure(figsize=(6, 4))  # Set figure size
        data = district_barriers[district_barriers['district'] == district]

        # Create a bar plot
        ax = sns.barplot(x='count', y='barriers_to_access', data=data, palette='bright')

        # Add data labels
        for container in ax.containers:
            ax.bar_label(container, fmt='%d', label_type='edge')

        # Set title and labels
        plt.title(f'Barriers to Access in {district} District', fontsize=16)
        plt.xlabel('Count', fontsize=14)
        plt.ylabel('Barrier Type', fontsize=14)

        # Display the plot
        plt.tight_layout()
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (adjust the file path as necessary)
df = pd.read_csv("C:/Users/alhas/Desktop/Jupiter Lab Project/SHS-Data.csv")

# Print the columns of the DataFrame to check for available satisfaction columns
print("DataFrame Columns:", df.columns.tolist())

# List of satisfaction columns
satisfaction_columns = [
    'cleanliness_satisfaction', 
    'staff_attitude_satisfaction', 
    'medicine_availability_satisfaction', 
    'waiting_time_satisfaction', 
    'quality_satisfaction', 
    'communication_satisfaction', 
    'feedback_satisfaction',
    'overall_satisfaction'
]

# Assuming a satisfaction scale from 1 to 5
max_score = 5

# Check if all satisfaction columns exist in the DataFrame
missing_columns = [col for col in satisfaction_columns if col not in df.columns]
if missing_columns:
    print(f"Warning: The following columns are missing from the DataFrame: {missing_columns}")
else:
    # Group the data by 'district' and calculate the mean satisfaction score for each district
    district_satisfaction = df.groupby('district')[satisfaction_columns].mean()

    # Calculate satisfaction rates in percentage for each district
    district_satisfaction_percent = district_satisfaction / max_score * 100

    # Reset index for easier plotting
    district_satisfaction_percent.reset_index(inplace=True)

    # Melt the DataFrame for easier plotting
    melted_df = district_satisfaction_percent.melt(id_vars='district', 
                                                    var_name='satisfaction_metric', 
                                                    value_name='satisfaction_percentage')

    # Get unique districts for separate plotting
    unique_districts = melted_df['district'].unique()

    # Create a plot for each district
    for district in unique_districts:
        # Filter data for the current district
        district_data = melted_df[melted_df['district'] == district]
        
        plt.figure(figsize=(10, 6))  # Set the figure size

        # Set a bright color palette for the bar plot
        bright_colors = sns.color_palette("bright", len(satisfaction_columns))

        # Create the horizontal bar plot for the current district
        sns.barplot(data=district_data, y='satisfaction_metric', x='satisfaction_percentage', 
                    palette=bright_colors)

        # Add titles and labels
        plt.title(f'Participant Satisfaction Rates (in %) for {district}', fontsize=16)
        plt.xlabel('Satisfaction Rate (%)', fontsize=14)
        plt.ylabel('Satisfaction Metrics', fontsize=14)

        # Add percentage labels on the bars
        for p in plt.gca().patches:
            plt.annotate(f'{p.get_width():.2f}%', 
                         (p.get_width(), p.get_y() + p.get_height() / 2), 
                         ha='left', va='center', 
                         fontsize=10, color='black', 
                         xytext=(5, 0),  # 5 points horizontal offset
                         textcoords='offset points')

        # Adjust layout to prevent clipping of tick labels
        plt.tight_layout()  
        
        # Save the plot as a PNG file for each district
        plt.savefig(f'participant_satisfaction_{district}.png', dpi=300, bbox_inches='tight')
        print(f"Plot saved as 'participant_satisfaction_{district}.png'")

        # Show the plot
        plt.show()
```

## Strengths and Areas of Improvement

```python
import pandas as pd

# Load your dataset (adjust the file path as necessary)
df = pd.read_csv("C:/Users/alhas/Desktop/Jupiter Lab Project/SHS-Data.csv")

# List of satisfaction columns
satisfaction_columns = [
    'cleanliness_satisfaction', 
    'staff_attitude_satisfaction', 
    'medicine_availability_satisfaction', 
    'waiting_time_satisfaction', 
    'quality_satisfaction', 
    'communication_satisfaction', 
    'feedback_satisfaction',
    'overall_satisfaction'
]

# Group the data by 'district' and calculate the mean satisfaction score for each district
mean_scores = df.groupby('district')[satisfaction_columns].mean()

# Calculate overall satisfaction score by taking the mean of the mean scores
overall_score = mean_scores.mean().mean()
print(f"\nOverall satisfaction score: {overall_score:.2f}")

# Identify strengths and areas for improvement
strengths = {}
areas_for_improvement = {}

# Set thresholds for identifying strengths and areas for improvement
strength_threshold = 4.0  # Adjust this threshold as necessary
improvement_threshold = 3.0  # Adjust this threshold as necessary

for column in satisfaction_columns:
    mean_score = mean_scores[column].mean()
    if mean_score >= strength_threshold:
        strengths[column] = mean_score
    elif mean_score <= improvement_threshold:
        areas_for_improvement[column] = mean_score

# Print strengths and areas for improvement
print("\nStrengths:")
for metric, score in strengths.items():
    print(f"{metric}: {score:.2f}")

print("\nAreas for Improvement:")
for metric, score in areas_for_improvement.items():
    print(f"{metric}: {score:.2f}")
```
